```python
import math
import numpy as np
```

```python
x = np.array([200, 17])
```

```python
W = np.array([[1, -3, 5], 
              [-2, 4, -6]])
```

```python
b = np.array([-1, 1, 2])
```

```python
def sigmod(z):
    return 1 / (1 + math.exp(-z))
```

```python
def dense(x, W, b):
    units = W.shape[1]
    out = np.zeros(units)
    for j in range(units):
        w = W[:,j]
        z = np.dot(w, x) + b[j]
        print(z)
        out[j] = sigmod(z)
    print(out)
    return out
```

```python
a_out = dense(x, W, b)
```
